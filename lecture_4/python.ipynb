```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv("../../data/heart_attack_prediction_dataset.csv")

df.info()
```

```python
# All the countries in the dataset
df["Country"].unique()
```

```python
# Balance of countries
df.groupby("Country")["Patient ID"].count().plot.barh()
```

```python
# Balance between Sex
df.groupby("Sex")["Patient ID"].count().plot.barh()
```

```python
# Balance of age
df["Age"].hist()
```

```python
# Is there a relationship between family history and heart disease?

sb.countplot(df, x="Family History", hue="Heart Attack Risk")
plt.title("Heart Attack Risk per Family History")
```

```python
# Q: If you are younger and have a heart attack it is more likely to be familial?
sb.histplot(df, x="Age", hue="Heart Attack Risk", multiple="stack", cbar=True)
```

```python
# Is there a relationship between high blood fats and heart attack?

df.info()
sb.boxplot(df, x="Heart Attack Risk", y="Triglycerides")
```

```python
# Heart attack frequency per country
p = sb.countplot(df, x="Country", hue="Heart Attack Risk")
_= p.set_xticklabels(p.get_xticklabels(), rotation=90)
```
