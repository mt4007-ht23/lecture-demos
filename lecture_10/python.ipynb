```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

# Simple Scraper

```python
URL = "https://www.worldometers.info/world-population/population-by-country/"

res = requests.get(URL)

html = BeautifulSoup(res.content)
```

```python
table = html.find("table", id="example2")
header = table.find("thead") 
body = table.find("tbody")

columns=[]
for variable in header.find_all("th"):
    columns.append(variable.get_text())

data = []

for row in body.find_all("tr"):

    row_data = []
    for val in row.find_all("td"):
        row_data.append(val.get_text())

    data.append(row_data)
```

```python
pd.DataFrame(data, columns=columns)
```

# Scraping ICA

```python
URL = "https://handlaprivatkund.ica.se/stores/1003422/categories"

res = requests.get(URL)

html = BeautifulSoup(res.content)
```

```python
def product_parser(product_html):
    """
    Input: Product HTML
    Output: Name, Price of product
    """
    name = product_html.find("a", attrs={"data-test": "fop-product-link"}).get_text()
    price = product_html.find("strong", attrs={"data-test": "fop-price"}).get_text()

    return name, price

products = html.find_all("div", attrs={"data-test": "fop-body"})


data = []
for prod in products:
    name, price = product_parser(prod)

    data.append((name,price))
```

```python
pd.DataFrame(data, columns=["product_name", "product_price"])
```
