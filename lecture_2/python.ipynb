```python
import pandas as pd
```

```python
df = pd.read_csv("../../data/Winter_medals2022-11-03.csv")

df.head(5)
```

```python
# SELECTING
df_select = df.loc[:,["Country", "Gold", "Silver", "Bronze"]]
```

```python
# GROUPING
df_grouping = df_select.groupby("Country").sum()
```

```python
# MUTATE
df_grouping.loc[:, ["Total"]] = df_grouping.apply(lambda x: sum(x), axis=1)
```

```python
# SORTING
df_grouping.sort_values("Total", ascending=False).head(10)
```

```python
#FILTERING
df_grouping.loc[(df_grouping["Gold"] > 10) & (df_grouping["Silver"] >= 30)]
```

```python
df_grouping.sort_values("Gold", ascending=False).head(5).mean(axis=0)
```
