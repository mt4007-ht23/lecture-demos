```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
country = pd.read_csv("../../data/country_data.csv")
population = pd.read_csv("../../data/population2022.csv")
covid = pd.read_csv("../../data/covid.csv")
```

```python
country.drop(["alpha-2", "country-code", "iso_3166-2",
             "region-code", "sub-region-code", "intermediate-region-code"], axis=1, inplace=True)
```

```python
country.loc[~country['intermediate-region'].isna(), :]
```

```python
country['intermediate-region'].fillna(country['sub-region'], inplace=True)
```

```python
country.dropna(inplace=True)
```

```python
df = country.merge(population, left_on="alpha-3", right_on="country-iso", how="inner")
#df = df.merge(covid, on="country-iso")
#df.loc[df.isna().any(axis=1), :]
```

```python
df["alpha-3"].fillna(df["country-iso"],inplace=True)
```

```python
df["name_x"].fillna(df["name_y"],inplace=True)
```

```python
df = df.merge(covid, left_on="alpha-3", right_on="country-iso", how="outer")
```

```python
df.drop(["name_y", "country-iso_x", "country-iso_y"],axis=1, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.rename({"name_x": "name"}, axis=1, inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
r = df.groupby("region")[["confirmed", "deaths"]].sum()
r["mr"] = 100*r["deaths"]/r["confirmed"]
r.sort_values("mr", ascending=False)
```

```python
sr = df.groupby("sub-region")[["confirmed", "deaths"]].sum()
sr["mr"] = 100*sr["deaths"]/sr["confirmed"]
sr.sort_values("mr", ascending=False)
```

```python
ir = df.groupby("intermediate-region")[["confirmed", "deaths"]].sum()
ir["mr"] = 100*ir["deaths"]/ir["confirmed"]
ir.sort_values("mr", ascending=False)
```

```python
r = df.groupby("region")[["confirmed", "population-2022"]].sum()
r["mr"] = 100*r["confirmed"]/r["population-2022"]
r.sort_values("mr", ascending=False)
```
